```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = {
    "I‘m happy.",
    "She exercises every morning.",
    "His dog barks loudly.",
    "My school starts at 8:00.",
    "We always eat dinner together.",
    "They take the bus to work.",
    "He doesn’t like vegetables.",
    "I don’t want anything to drink.",
    "This little black dress isn’t expensive. ",
    "Those kids don’t speak English."
}
#unqual size
voc_size = 10000
oh = [one_hot(s,voc_size) for s in sent]
```

```python
oh
```

```python
#word embedding representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
import pandas as pd
```

```python
sent_len=6
embedded_docs=pad_sequences(oh,padding='pre',maxlen=sent_len)
```

```python
embedded_docs
```

```python
dim = 6

model =Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_len))
model.compile('adam','mse')
```

```python
model.predict(embedded_docs)
```
